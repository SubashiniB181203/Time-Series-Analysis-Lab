***7. Implement program for decomposing time series data into trend and seasonality***

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Load dataset
def load_data(filepath, date_column):
    df = pd.read_csv(filepath, parse_dates=[date_column], index_col=date_column)
    return df
```

```python
# Decompose time series
def decompose_time_series(df, column, model='additive', period=None):
    decomposition = seasonal_decompose(df[column], model=model, period=period)
    return decomposition
```

```python
# Plot decomposition results
def plot_decomposition(decomposition):
    fig, axes = plt.subplots(4, 1, figsize=(10, 8), sharex=True)
    colors = ['blue', 'green', 'red', 'purple']  # Different colors for each plot

    decomposition.observed.plot(ax=axes[0], title='Observed', color=colors[0])
    decomposition.trend.plot(ax=axes[1], title='Trend', color=colors[1])
    decomposition.seasonal.plot(ax=axes[2], title='Seasonality', color=colors[2])
    decomposition.resid.plot(ax=axes[3], title='Residuals', color=colors[3])

    plt.tight_layout()
    plt.show()
```

```python
# Example usage
if __name__ == "__main__":
    filepath = '/content/sample_data/time_series_data.csv'  # Update with actual file path
    date_column = 'DATE'  # Adjust as needed
    value_column = 'IPG2211A2N'  # Adjust as needed
    df = load_data(filepath, date_column)
    decomposition = decompose_time_series(df, value_column, period=12)
    plot_decomposition(decomposition)
```
