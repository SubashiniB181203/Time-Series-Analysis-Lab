***2. Visualizing Time Series Data***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/sample_data/time_series_data.csv",parse_dates=['DATE'],index_col='DATE')
print(data.head())
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['IPG2211A2N'],label='Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Line Plot of Time Series Data')
plt.legend()
plt.show()
```

```python
monthly_data=data['IPG2211A2N'].resample('M').mean()
plt.figure(figsize=(12,6))
plt.plot(monthly_data,label='Monthly Average')
plt.title('Monthly Trends')
plt.xlabel('Date')
plt.ylabel('Average Value')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
data['IPG2211A2N'].hist(bins=30)
plt.title('Histogram of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=data,x='IPG2211A2N')
plt.title('Box Plot of Values')
plt.show()
```
